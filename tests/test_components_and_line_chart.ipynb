```python
import pandas as pd
import numpy as np
from quantfin.macro.charts import draw_components_and_line_chart
import seaborn as sns
```

```python
dates = pd.date_range(start='1990-01',end='2023-12',freq="QE")
data_ = np.random.normal(size=(len(dates),3))
df = pd.DataFrame(data_, columns=['a','b','c'], index=dates)
df = df.cumsum()
df.columns.name = 'test_columns'
df.index.name = "quarter"
```
