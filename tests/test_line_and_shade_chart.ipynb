```python
import pandas as pd
import numpy as np
from quantfin.macro.charts import draw_line_and_shade_chart
import seaborn as sns
```

```python
dates = pd.date_range(start='1990-01',end='2023-12',freq="QE")
data_ = np.random.normal(size=(len(dates),3))
df = pd.DataFrame(data_, columns=['a','b','c'], index=dates)
df = df.cumsum()
df.columns.name = 'test_columns'
df.index.name = "quarter"
```

```python
df.head()
```

```python
shading = pd.Series(False, index=dates)
shading.loc['2020'] = True
shading.loc['2010-01':'2010-09'] = True
```

```python
sns.set_style("whitegrid")
```

```python
fig, ax = draw_line_and_shade_chart(df, shading, linewidth=3)
```

```python
h_shading = pd.Series([False, True, True, False], index=pd.Index([0., 2.5, 5.0, 7.], name='idx'), name='Vshade')
draw_line_and_shade_chart(df, h_shading, v_h="horizontal");
```
